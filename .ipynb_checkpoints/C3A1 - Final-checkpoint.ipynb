```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from pyod.models.knn import KNN
from sklearn.feature_selection import mutual_info_regression
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error

SEED = (1995)
np.random.seed(SEED)

sns.set_style('whitegrid')

pd.set_option('display.max_rows', 80)
pd.set_option('display.max_columns', 80)
```

```python
full_sales = pd.read_csv('train.csv')
print('Linhas:', full_sales.shape[0], '| Colunas:', full_sales.shape[1])
full_sales.head()
```

```python
#Treino e Validação

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y
   , train_size=0.70
    , test_size=0.30
    , random_state=SEED
)

#Listando features categóricas, ou seja, todas que possuam texto:
print('Features Categóricas:', categorical_features)
```

```python
#Listando good e bad labels:

good_labels_categorical_features = [
    col for col in categorical_features
    if set(X_valid[col]).issubset(set(X_train[col]))
]

bad_labels_categorical_features = list(
    set(categorical_features) - set(good_labels_categorical_features)
)

# Listando os Resultados
print('Good Labels:', good_labels_categorical_features)
print('\nBad Labels:', bad_labels_categorical_features)
```

```python
# Retirando as Bad Labels
good_labels_X_train = X_train.drop(bad_labels_categorical_features, axis=1).copy()
good_labels_X_valid = X_valid.drop(bad_labels_categorical_features, axis=1).copy()

print('Colunas:', good_labels_X_train.shape[1])
good_labels_X_train.head()
```

```python
#Verificando a existência de dados nulos

good_labels_X_train.isnull().sum().sort_values(ascending=False)
```

```python
good_labels_X_valid.isnull().sum().sort_values(ascending=False)
```

```python
# Listando todas as Features Categóricas depois de remover as Bad Labels
categorical_features = [
    col for col in good_labels_X_train.columns
    if good_labels_X_train[col].dtype == 'object'
]

print('Features Categóricas:', categorical_features)
```

```python
# Armazenando os valores únicos delas
categorical_features_unique_values = { }

for column in categorical_features:
    categorical_features_unique_values[column] = list(enumerate(good_labels_X_train[column].unique()))
```

```python
# Aplicando o Encoding
for feature in categorical_features:
    for index in range(len(categorical_features_unique_values[feature])):
        good_labels_X_train[feature].replace(
            categorical_features_unique_values[feature][index][1]
            , categorical_features_unique_values[feature][index][0]
            , inplace=True
        )
        
        good_labels_X_valid[feature].replace(
            categorical_features_unique_values[feature][index][1]
            , categorical_features_unique_values[feature][index][0]
            , inplace=True
        )
```

```python
# Checando o tipo de dados novamente (Dataset de Treino)
good_labels_X_train.dtypes
```

```python
# Checando o tipo de dados novamente (Dataset de Validação)
good_labels_X_valid.dtypes
```

```python
# Definindo a porcentagem de outliers e o modelo de classificação
outliers_fraction = 0.07
clf_name='KNN'
clf = {
    clf_name : KNN(
                contamination=outliers_fraction
                , method='largest'
                , n_neighbors=5
                , n_jobs=4
              )
}
```

```python
# Alterando todos os valores nulos por zeros em datasets temporários
features_with_missing_values = [
    'PoolQC' , 'MiscFeature', 'Alley'
    , 'Fence', 'FireplaceQu', 'LotFrontage'
    , 'GarageYrBlt', 'GarageCond', 'GarageType'
    , 'GarageFinish', 'BsmtExposure', 'BsmtFinType2'
    , 'BsmtCond', 'BsmtFinType1', 'BsmtQual'
]

good_labels_X_train_temp = good_labels_X_train.fillna(0).copy()
good_labels_X_valid_temp = good_labels_X_valid.fillna(0).copy()
```

```python
# Treinando o modelo e realizando predições
clf[clf_name].fit(good_labels_X_train_temp)

# Armazenando as Labels Binárias (0: inliers, 1: outliers)
X_train_predictions = clf[clf_name].labels_
X_valid_predictions = clf[clf_name].predict(good_labels_X_valid_temp)
```

```python
# Armazenando todos os índices das linhas que contém outliers
outliers_indexes_X_train = [
    index for index in range(0, len(X_train_predictions))
    if X_train_predictions[index] == 1
]

outliers_indexes_X_valid = [
    index for index in range(0, len(X_valid_predictions))
    if X_valid_predictions[index] == 1
]
```

```python
# Criando um dataset apenas com os outliers
outlier_rows_X_train = good_labels_X_train.iloc[outliers_indexes_X_train]
outlier_rows_X_valid = good_labels_X_valid.iloc[outliers_indexes_X_valid]

percentage_outliers_X_train = len(outliers_indexes_X_train) * 100 / len(good_labels_X_train)
percentage_outliers_X_valid = len(outliers_indexes_X_valid) * 100 / len(good_labels_X_valid)

print('Porcentagem de Outliers - Dataset de Treino:', percentage_outliers_X_train)
print('Porcentagem de Outliers - Dataset de Validação:', percentage_outliers_X_valid)
```

```python
# Retirando os outliers (X - Datasets de Treino e Validação)
good_labels_X_train.drop(good_labels_X_train.index[outliers_indexes_X_train], inplace=True)
good_labels_X_valid.drop(good_labels_X_valid.index[outliers_indexes_X_valid], inplace=True)
```

```python
# Retirando os outliers (X - Datasets Temporários de Treino e Validação)
good_labels_X_train_temp.drop(good_labels_X_train_temp.index[outliers_indexes_X_train], inplace=True)
good_labels_X_valid_temp.drop(good_labels_X_valid_temp.index[outliers_indexes_X_valid], inplace=True)
```

```python
# Retirando os outliers (y - Datasets de Treino e Validação)
y_train.drop(y_train.index[outliers_indexes_X_train], inplace=True)
y_valid.drop(y_valid.index[outliers_indexes_X_valid], inplace=True)
```

```python
#FEATURE ENGINEERING:
#MUTUAL INFORMATION - Verificar as Features mais importantes

# Função para Visualizar os scores da MI
def plot_mi_scores(scores):
    """
    Exibe um gráfico com os scores da MI em order crescente
    """
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information - Scores")
    plt.show()
```

```python
# Armazenando todas as Features discretas (números inteiros) e contínuas (números decimais)
discrete_features = [
    feature for feature in good_labels_X_train.columns
    if good_labels_X_train[feature].dtype == 'int64'
]

continuous_features = [
    feature for feature in good_labels_X_train.columns
    if good_labels_X_train[feature].dtype == 'float64'
]

# Criando datasets temporários das Features discretas e contínuas
discrete_features_list = good_labels_X_train[discrete_features].dtypes == 'int64'
continuous_features_list = good_labels_X_train[continuous_features].dtypes == 'float64'
```

```python
# Calculando os scores MI
mi_scores_discrete_features = mutual_info_regression(good_labels_X_train_temp[discrete_features], y_train, discrete_features=discrete_features_list, random_state=SEED)
mi_scores_discrete_features = pd.Series(mi_scores_discrete_features, name='MI Scores - Discrete Features', index=discrete_features)
mi_scores_discrete_features = mi_scores_discrete_features.sort_values(ascending=False)

mi_scores_continuous_features = mutual_info_regression(good_labels_X_train_temp[continuous_features], y_train, discrete_features=continuous_features_list, random_state=SEED)
mi_scores_continuous_features = pd.Series(mi_scores_continuous_features, name='MI Scores - Continuous Features', index=continuous_features)
mi_scores_continuous_features = mi_scores_continuous_features.sort_values(ascending=False)

# Exibindo os resultados
plt.figure(dpi=100, figsize=(15, 20))
plot_mi_scores(mi_scores_discrete_features)

plt.figure(dpi=100, figsize=(10, 10))
plot_mi_scores(mi_scores_continuous_features)
```

```python
#desconsiderando score menor que 0.11

features_to_drop = list(mi_scores_discrete_features[mi_scores_discrete_features < 0.11].index)
good_labels_X_train.drop(features_to_drop, axis=1, inplace=True)
good_labels_X_valid.drop(features_to_drop, axis=1, inplace=True)
```

```python
print('Colunas:', good_labels_X_train.shape[1])
good_labels_X_train.head()
```

```python
#Clustering
# Verificando a melhor quantidade de clusters com o método elbow e WCSS
wcss = []

for i in range(1,7):
    kmeans = KMeans(
        n_clusters=i
        , init='k-means++'
        , max_iter=300
        , n_init=10
        , random_state=SEED
    )
    kmeans.fit(good_labels_X_train_temp)
    wcss.append(kmeans.inertia_)
    
# Plotting the results
plt.plot(range(1,7), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

#Há curvas em 2,3,4 e 5 clusters, estaremos utilizando 4
```

```python
kmeans = KMeans(
    n_clusters=4
    , init='k-means++'
    , max_iter=300
    , n_init=10
    , random_state=SEED
)

good_labels_X_train_temp.insert(loc=0, column='Cluster', value=kmeans.fit_predict(good_labels_X_train_temp))
good_labels_X_train_temp.head()
```

```python
# Verificando o Padrão dos Clusters com as Features 'YearBuilt' e 'LotArea'
sns.relplot(
    x='YearBuilt', y='LotArea', hue='Cluster', data=good_labels_X_train_temp, height=6, palette='rocket'
);
```

```python
#Podemos concluir que:
# Cluster 0 -> LotArea entre 0 - 6,000 - Lotes pequenos com uma concetração maior entre 1980 e 2000
# Cluster 1 -> LotArea entre 10,000 - 12,000 - Lotes medianos bem distribuídos ao longo do tempo
# Cluster 2 -> LotArea entre 7,000 - 10,000 - Lotes médios com concentração maior entre 1940 e 1980
# Cluster 3 -> LotArea entre 13,000+ - Lotes grandes concentrados principalmente entre 1950 e 2000
```

```python
#Inserindo os clusters nos datasets originais: good_labels_X_train e good_labels_X_valid.

good_labels_X_train.insert(loc=0, column='Cluster', value=good_labels_X_train_temp['Cluster'])

good_labels_X_valid_temp.insert(loc=0, column='Cluster', value=kmeans.fit_predict(good_labels_X_valid_temp))
good_labels_X_valid.insert(loc=0, column='Cluster', value=good_labels_X_valid_temp['Cluster'])

good_labels_X_train.head()
```

```python
# Verificando o a quantidade de Features
print('Quantidade de Features:', good_labels_X_train.shape[1])
```

```python
rmse = lambda predictions, real_values: np.sqrt(mean_squared_error(predictions, real_values))
```

```python
# Pipelines de transformação numérica
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer #para preencher valores ausentes com constantes
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.compose import ColumnTransformer


#Os scalers serão utilizados para padronizar as váriaveis numéricas
# Transformador das Features Numéricas
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant'))
#    , ('scaler', StandardScaler())
#   ,  ('scaler', RobustScaler())
   ,  ('scaler', MinMaxScaler())
#   ,  ('scaler', Normalizer())
])

# Mesclando os Pré-processadores
preprocessor = ColumnTransformer(
    transformers=[('numerical', numerical_transformer, good_labels_X_train.columns)]
)
```

```python
from xgboost import XGBRegressor
```

```python
# Criando o modelo
xgb_model = XGBRegressor(
    objective='reg:squarederror'
    , n_estimators=100
    , learning_rate=0.10
    , colsample_bytree=0.70
    , max_depth=3
    , n_jobs=4
)

xgb_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])
```

```python
# Treinando o modelo e realizando previsões
xgb_X_train = xgb_pipeline.fit_transform(good_labels_X_train)
xgb_X_valid = xgb_pipeline.transform(good_labels_X_valid)

xgb_model.fit(
    xgb_X_train,
    y_train,
    # early_stopping_rounds=5,  # Desativado
    eval_set=[(xgb_X_valid, y_valid)],
    verbose=False
)

print('Treino Finalizado!')

xgb_predictions = xgb_model.predict(xgb_X_valid)
print('Previsões Finalizadas!')
```

```python
# RMSE
xgb_rmse = rmse(xgb_predictions, y_valid)
print('XGBoost Regressor RMSE: ', xgb_rmse)
```

```python
# Scores de Treino e Validação
print('Score de Treino: %.2f%%' % (xgb_model.score(xgb_X_train, y_train) * 100))
print('Score de Validação: %.2f%%' % (xgb_model.score(xgb_X_valid, y_valid) * 100))
```

```python
#Realizando Permutation Importance para descobrir quais Features são mais importantes

sorted_permutation_importance = xgb_model.feature_importances_.argsort()

plt.figure(figsize=(15,20))
plt.title('XGBoost - Feature Importance')
plt.xlabel("XGBoost Feature Importance")
plt.barh(good_labels_X_valid.columns[sorted_permutation_importance], xgb_model.feature_importances_[sorted_permutation_importance])

plt.show();
```

```python
#Podemos ver que o XGBoost considerou GarageCars, OverallQual e, de certa forma, BsmtQual como Features mais importantes.
#Ele tamvém considerou 2ndFlrSF e outras abaixo as Features menos importantes
```
